```python
import pandas as pd
%matplotlib inline
```

```python
sci = pd.read_csv('sci.csv')
sci.head()
```

```python
sci.set_index('Date', inplace=True)
sci.head()
```

```python
import plotly.express as px
fig = px.line(sci, x=sci.index, y='Adj Close')
fig.show()
```

### a)

```python
total_return = (sci['Adj Close'][len(sci)-1]-sci['Adj Close'][0])/sci['Adj Close'][0]
total_return
```

```python
mean_annualized_return = total_return**(1/20)-1
mean_annualized_return
```

### b)

```python
daily_returns = sci.pct_change()[1::]
daily_returns.head()
```

```python
fig = px.line(daily_returns, x=daily_returns.index, y='Adj Close')
fig.show()
```

```python
# Annualized volatility is daily volatility * sqrt(252)
len(daily_returns)
```

```python
annual_vol = daily_returns.std()*(252**0.5)
annual_vol
```

### c)

```python
fig = px.histogram(daily_returns, x="Adj Close")
fig.show()
```

```python
import scipy.stats as stats
```

```python
x = daily_returns['Adj Close']
kurtosis = stats.kurtosis(x)
print("kurtosis is {}".format(kurtosis))
skew = stats.skew(x)
print("skewness is {}".format(skew))
```

Therfore, since the kurtosis is 3, the daily returns of the SCI is leptokurtic. While the skew is lightly smaller compared to that of a normal distribution. **There is an evidence of a fat tail. (high kurtosis)**.

### d)

```python
import numpy as np
```

```python
# pd.plotting.autocorrelation_plot(daily_returns['Adj Close'])
lags = []
auto_corrs = []
for i in range(10):
    lags.append(i+1)
    auto_corr = pd.Series(daily_returns['Adj Close']).autocorr(lag = i+1)
    auto_corrs.append(auto_corr)
print(np.mean(auto_corrs))
fig = px.line(x = lags,y = auto_corrs)
fig.show()
```

According to the auto correlation with lag from 1 to 10, the mean of autocorrelation is close to 0. 
**CSI does not satisfy but very close to efficient market hypothesis.**
