```python
import plotly.graph_objects as go
import pandas as pd

# df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv")
df = pd.read_csv('Calculated_IVIX.csv')
fig = go.Figure()
fig.add_trace(go.Scatter(
                x=df['DateTime'],
                y=df['true_ivix'],
                name="True_ivix",
                line_color='red',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=df['DateTime'],
                y=df['calculated_ivix'],
                name="calculated_ivix",
                line_color='blue',
#                 line_color='dimgray',
                opacity=0.8))

# # Use date string to set xaxis range
# fig.update_layout(xaxis_range=['2016-07-01','2016-12-31'],
#                   title_text="Manually Set Date Range")
fig.show(render='png')
```

```python
from scipy.stats.mstats import winsorize
df['winsorized_civix'] = winsorize(df['calculated_ivix'],limits=[0.1, 0])

fig.add_trace(go.Scatter(
                x=df['DateTime'],
                y=df['winsorized_civix'],
                name="winsorized_civix",
                line_color='orange',
#                 line_color='dimgray',
                opacity=0.8))
fig.show()
```
