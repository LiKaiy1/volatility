### a)

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
import matplotlib
matplotlib.rc('xtick', labelsize=40) 
matplotlib.rc('ytick', labelsize=40) 
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
from statsmodels.tsa.stattools import acf, pacf
```


From the ACF plot, we can see that there are some (but small) autocorrelations for the firts few lags. We conclude that: __Although autocorrelations exist in the data and the efficient market hypothesis is not strictly valid, but the market is very close to satisfy the efficient market hypothesis.__

## 2.

For daily window:\
Annualized Historical Volatility is\
$$
\sigma = \sqrt{252}\times\sigma_{daily}
$$

```python
daily_window_annual_volatility = sci_daily_returns*(252**0.5)
# daily_window_annual_volatility.head()
```

For montly window:

```python
# First: Resample monthly returns.
sci_monthly_price = sci['Adj Close'].resample('M').last()
sci_monthly_price = pd.DataFrame(sci_monthly_price)
```

```python
# Second: Insert the first month data
sci_monthly_price = sci_monthly_price.append(pd.DataFrame(sci['Adj Close'].iloc[sci.index == pd.to_datetime('2000-01-03',format='%Y-%m-%d')]))
sci_monthly_price.sort_index(inplace=True) 
```

```python
# Then: Calculate the monthly log return.
sci_monthly_returns = np.log(sci_monthly_price['Adj Close']).diff()[1::]
sci_monthly_returns = pd.DataFrame(sci_monthly_returns)
sci_monthly_returns['monthly.return'] = sci_monthly_returns['Adj Close']
sci_monthly_returns = sci_monthly_returns[['monthly.return']]
```

For monthly window:\
Annualized Historical Volatility is\
$$
\sigma = \sqrt{252}\times\sigma_{month}
$$

```python
monthly_window_annual_volatility = sci_monthly_returns*(12**0.5)
```

For annually window:\
Annualized Historical Volatility is\
$$
\sigma = \sqrt{1}\times\sigma_{year}
$$

```python
annually_window_annual_volatility = sci_annual_returns
```

### a)

```python
print("The difference between maximum and minimum of daily window volatility is {}."
      .format(float(daily_window_annual_volatility.max())
              -float(daily_window_annual_volatility.min())))

print("The difference between maximum and minimum of monthly window volatility is {}."
      .format(float(monthly_window_annual_volatility.max())
             -float(monthly_window_annual_volatility.min())))

print("The difference between maximum and minimum of annually volatility is {}"
      .format(float(annually_window_annual_volatility.max())
             -float(annually_window_annual_volatility.min())))
```

Therefore, __daily window has the greatest difference between highest and lowest.__

```python
print("The mean of daily volatility is {}.".format(float(daily_window_annual_volatility.mean())))
print("The mean of monthly volatility is {}.".format(float(monthly_window_annual_volatility.mean())))
print("The mean of annually volatility is {}.".format(float(annually_window_annual_volatility.mean())))
print("----------------------------------------------------")
print("The standard deviation of daily volatility is {}.".format(float(daily_window_annual_volatility.std())))
print("The standard deviation of monthly volatility is {}.".format(float(monthly_window_annual_volatility.std())))
print("The standard deviation of annually volatility is {}.".format(float(annually_window_annual_volatility.std())))
print("----------------------------------------------------")
print("The kurtosis of daily volatility is {}.".format(float(kurtosis(daily_window_annual_volatility['daily.return']))))
print("The kurtosis of monthly volatility is {}.".format(float(kurtosis(monthly_window_annual_volatility['monthly.return']))))
print("The kurtosis of annually volatility is {}.".format(float(kurtosis(annually_window_annual_volatility['ann.return']))))
```

__The annually window generates highest mean and standard deviation, while monthly window has the highest kurtosis.__
