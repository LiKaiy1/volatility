 Before we start everything, we read the data.

```r
bac_price <- read.csv('bac19.csv',head = T)
bac <- bac_price$bac
date <- bac_price$date
bac_price <- data.frame(
  DATE = as.Date.factor(date),
  price = bac
)
```

```r
library(ggplot2)
library(dplyr)

p <- ggplot(bac_price, aes(x=DATE, y=price)) +
  geom_line() + 
  xlab("")
p
```
